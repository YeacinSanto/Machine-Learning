```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "h" : [150,160,170,180,190],
    "w" : [50,60,70,80,90]
})
df
```

```python
m = df.mean()
# print(m)
sd = df.std()
# print(sd)
z = (df-m)/sd
z.round(2)
```

```python
z.plot(kind="bar")
plt.title("Z scaling")
plt.show()
```

```python
# min mmax scaling
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "h" : [150,160,170,180,190],
    "w" : [50,60,70,80,90]
})
df
```

```python
mn = df.min()
mx = df.max()
rg = mx-mn
mn,mx,rg
```

```python
ss = (df - mn)/rg
ss
```

```python
ss.plot(kind="bar")
plt.title("min max scaling")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "h" : [150,160,170,180,300],
    "w" : [50,60,70,80,200]
})
df
```

```python
md = df.median()
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3-q1
md,q1,q3,iqr
```

```python
ct = (df-md)/iqr
ct
```

```python
ct.plot(kind="bar")
plt.title("robust scaling")
plt.show()
```

```python
out = pd.concat([df,ct],axis=1)
out
```

```python
import pandas as pd

df = pd.DataFrame({
    "id" : [1,2,3,4],
    "color" : ["red","blue","green","red"],
    "size" : ["Small","Medium","Large","Medium"],
    "price" : [10,12,15,11]
})
df
```

```python
d_color = pd.get_dummies(df["color"], prefix="color", dtype=int)
d_color
```

```python
df_encoded = pd.concat([df,d_color],axis=1)
df_encoded
```
